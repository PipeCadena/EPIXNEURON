```python
import pandas as pd

# Especifica la ruta o URL del archivo CSV que deseas cargar
archivo_csv = 'combined_sweeps_final.csv'

# Define los nombres de las columnas personalizados
nombres_columnas = ["Paciente", "Tiempo", "Corriente", "Voltaje"]

# Utiliza pandas para cargar el archivo CSV en un DataFrame y asignar los nombres de las columnas
df = pd.read_csv(archivo_csv, names=nombres_columnas, header=None)
df
```

```python
import pandas as pd
import numpy as np

# Supongamos que tienes un DataFrame df con las columnas "Paciente", "Tiempo", "Corriente" y "Voltaje"

# Agrupa los datos por paciente y nivel de corriente
grupos = df.groupby(["Paciente", "Corriente"])

# Prepara una estructura para almacenar los resultados
resultados = []

# Define una funciÃ³n para calcular la frecuencia de disparo
def calcular_frecuencia_disparo(voltajes, umbral=40):
    spikes = np.where(voltajes > umbral)
    print(spikes)
    tiempo_total = len(voltajes)  # DuraciÃ³n total del registro
    frecuencia = len(spikes) / (tiempo_total)  # Frecuencia en Hz (suponiendo que el tiempo estÃ© en milisegundos)
    return frecuencia

# Itera a travÃ©s de los grupos
for (paciente, corriente), grupo in grupos:
    voltajes = grupo["Voltaje"]  # Convierte la columna de voltajes en una lista
    # Calcula la frecuencia de disparo para esta corriente
    frecuencia = calcular_frecuencia_disparo(voltajes)
    
    resultados.append({"Paciente": paciente, "Corriente": corriente, "Voltajes": voltajes, "Frecuencia": frecuencia})

# Ahora tienes una lista de resultados, donde cada elemento es un diccionario con el paciente, la corriente, la lista de voltajes y la frecuencia de disparo
```

```python
import matplotlib.pyplot as plt

# Supongamos que tienes una lista de resultados que contiene diccionarios con "Paciente", "Corriente", "Voltajes" y "Frecuencia"
# Cada paciente tiene un identificador Ãºnico en la columna "Paciente"

# Crea un diccionario que mapea pacientes a colores
colores = {}
pacientes = set()
for resultado in resultados:
    paciente = resultado["Paciente"]
    pacientes.add(paciente)

# Asigna un color Ãºnico a cada paciente
paleta_colores = plt.cm.get_cmap('tab10', len(pacientes))

for i, paciente in enumerate(pacientes):
    colores[paciente] = paleta_colores(i)

# Prepara una figura para la grÃ¡fica
plt.figure(figsize=(10, 6))

# Itera a travÃ©s de los resultados y traza las curvas de frecuencia vs. corriente para cada paciente
for resultado in resultados:
    paciente = resultado["Paciente"]
    corriente = resultado["Corriente"]
    frecuencia = resultado["Frecuencia"]
    
    plt.plot(corriente, frecuencia, marker='o', linestyle='-', label=f"Paciente {paciente}", color=colores[paciente])

# AÃ±ade etiquetas y leyenda
plt.xlabel("Corriente (pA)")
plt.ylabel("Frecuencia de Disparo (Hz)")
plt.title("Curva de Frecuencia de Disparo vs. Corriente por Paciente")
plt.legend()
plt.grid(True)

# Muestra la grÃ¡fica
plt.show()

```
